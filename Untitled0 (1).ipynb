**Q1. What is Power BI and why is it used in businesses?**

**Ans-**

**What is Power BI **


*   Power BI is a business intelligence tool by Microsoft.

*   Helps visualize data and create interactive reports.

*   Supports multiple data sources and real-time analytics.

*   Used by analysts, managers, and decision-makers.

*   Enables faster insights compared to static Excel reports.

**Why Businesses Use Power BI**


*   **Data-driven decisions** → Provides real-time insights so leaders act on facts, not guesswork.

*   **Visualization** → Converts complex spreadsheets into easy-to-read dashboards.

*   **Collaboration** → Teams share reports securely across devices and apps.

*   **Automation** → Updates dashboards automatically with the latest data.

*   **Cost efficiency** → Reduces reliance on multiple analytics tools, lowering IT costs.
*   **AI integration** → Uses Copilot and machine learning to uncover trends and generate summaries.





**Q2. Name and explain the three main components of Power BI**

**Ans-**

**Three main components of Power BI is**



1.   **Power BI Desktop –** Build Reports

Power BI Desktop is a free desktop application where users connect data, clean it using Power Query, and build reports.



2.   **Power BI Service –** Publish & Share Reports

Power BI Service is a cloud-based platform (powerbi.com) used to publish, share, and collaborate on reports created in Power BI Desktop.


3.  **Power BI Mobile –** View Reports on Mobile

Power BI Mobile is a mobile app (Android & iOS) used to view and interact with    reports on smartphones and tablets.

**Q6.  Define Data View, Report View, and Model View. Explain the purpose of each view.**

**Ans-**



1.  **Data View**

*   **Purpose:** Shows the raw data tables that you have imported or transformed.

*   **Use:**

    *   Inspect rows and columns.
    *   Verify data types and values.
    *   Check if cleaning steps (like removing duplicates or handling nulls) worked correctly.

2.   **Report View**

*   **Purpose:** The main canvas where you design and build interactive reports.

*   **Use:**

    *   Add visuals (charts, tables, KPIs, slicers).
    *   Arrange layout and apply formatting.
    *   Create dashboards that communicate insights.

3.   **Model View**






*   **Purpose:** Displays relationships between different tables in your dataset.
*   **Use:**
    *   Define relationships (one-to-many, many-to-many).
    *   Manage how tables connect for accurate calculations.
    *   Ensure measures and visuals pull correct data.

**Q8.  Split Owner Name to create two new columns as First Name and Last Name**

**Ans-**

**To split the Owner_Name column into First Name and Last Name in Power Query:**

1.   Select the Owner_Name column.

2.   Go to Transform → Split Column → By Delimiter.

3.   Choose Space as the delimiter.

4.   This will create two new columns: Owner_Name.1 (First Name) and Owner_Name.2 (Last Name).
5.   Rename them to First Name and Last Name.


Example:



*   "Rahul Mehta" → First Name = Rahul, Last Name = Mehta
*   "Anita Sharma" → First Name = Anita, Last Name = Sharma
